Q2. Why *args and **kwargs is used in some functions? Create a function each for *args and **kwargs
to demonstrate their use.

Answer:*args allows a function to accept any number of positional arguments. The *args parameter collects all the positional arguments passed to the function into a tuple. This is useful when you want to pass a variable number of arguments without explicitly specifying them in the function definition
**kwargs allows a function to accept any number of keyword arguments. The **kwargs parameter collects all the keyword arguments passed to the function into a dictionary. This is useful when you want to pass a variable number of named arguments without explicitly specifying them in the function definition.
Example for *args and **kwargs

```python
def def_arg(*args):
    total=0
    for i in args:
        total=total+i
    return total
def_arg(1,5,8,6)
def_arg(8,8,5,7,2,1)
#Now creating **kwargs function
def def_kwarg(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}:{value}")
def_kwarg(Name="Syed Rahman Ur Lateef",age=22,city="hyderabad")
```

Q4. What is a generator function in python? Why yield keyword is used? Give an example of a generator
function.

Answer:Generator is used to create an iterator method by using yield keyword we can create an iterator of any variable is called generator

```python
# Example of Generator
l=[4,4,5,6,7,8,9,10,11,12,34]
l1=[]
for i in l:
    yield l1
    l1.append(i)
    print(l1)
```

Q5. Create a generator function for prime numbers less than 1000. Use the next() method to print the
first 20 prime numbers.

```python
def prime_generator():
    primes = []

    for num in range(2, 1000):
        is_prime = True
        for prime in primes:
            if num % prime == 0:
                is_prime = False
                break

        if is_prime:
            primes.append(num)
            yield num

# Create the prime number generator
prime_gen = prime_generator()

# Print the first 20 prime numbers
for _ in range(20):
    prime = next(prime_gen)
    print(prime)
```

Q6. Write a python program to print the first 10 Fibonacci numbers using a while loop.

```python
def get_fib(n):
    a, b = 0, 1
    count = 0
    while count < n:
        yield a
        c = a + b
        a = b
        b = c
        count += 1
fib=get_fib(10)
for i in range(10):
    print(next(fib))

```

Q8. Write a python program to check whether a given number is Palindrome or not using a while loop.

```python
def is_palindrome(num):
    original_num = num
    reverse_num = 0

    while num > 0:
        remainder = num % 10
        reverse_num = (reverse_num * 10) + remainder
        num = num // 10

    if original_num == reverse_num:
        return True
    else:
        return False

# Test the function
number = int(input("Enter a number: "))
if is_palindrome(number):
    print(number, "is a palindrome.")
else:
    print(number, "is not a palindrome.")
```
